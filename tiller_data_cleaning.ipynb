Since the waiter_id is under 2% and 38 in total number, we can replace it with mode.

```python
order_data['id_device'].nunique()
```

Since the id_device is 30 in total number, with one of the top enteries of data, so we can't replace it with mode rather remove all the indexes, where id_device is null.

```python
device_drop_index= order_data[order_data['id_device'].isnull()].index
device_drop_index
```

```python
order_data.drop(index= device_drop_index, inplace = True, axis= 0)
```

```python
order_data['dim_source'].value_counts()
```

```python
order_data['dim_source'].value_counts()/len(order_data)*100
```

Since the difference between Tiller iPAD and Tiller-order is very much, so we will fill the missing values with higher weightage of Tiller iPAD

```python
order_data['dim_source'].fillna('Tiller iPAD', axis= 0, inplace= True)
```

```python
#To standardize the entries of data in dim_source column

order_data['dim_source'].replace('tiller-order', 'Tiller-order', inplace= True)
```

```python
order_data.isnull().sum()
```

```python
order_data['date_opened']
```

```python
breakfast_start = datetime.strptime('00:00:00', '%H:%M:%S').time()
breakfast_end = datetime.strptime('11:59:59', '%H:%M:%S').time()

lunch_start = datetime.strptime('12:00:00', '%H:%M:%S').time()
lunch_end = datetime.strptime('17:59:59', '%H:%M:%S').time()

dinner_start = datetime.strptime('18:00:00', '%H:%M:%S').time()
dinner_end = datetime.strptime('23:59:59', '%H:%M:%S').time()
```

```python
order_data['time']= order_data['date_opened'].dt.time
```

```python
#Create a function to categorize breakfast, lunch and dinner.

def meal(time):
  if time < breakfast_end:
    return 'Breakfast'
  elif time < lunch_end:
    return 'Lunch'
  else:
    return 'Dinner'
```

```python
order_data['meal_types']= order_data['time'].apply(meal)
```

```python
order_data.head(15)
```

```python
#Exporting the cleaned order data set to csv

#order_data.to_csv('/content/drive/MyDrive/order_data_cleaned.csv')

order_data.to_csv('order_data_cleaned.csv', index= False)
```

```python
#Create a new function to translate and lowercase the text

def translate_and_lowercase(text):
    translator = GoogleTranslator(source='auto', target='en')  # Translate to English
    translated_text = translator.translate(text)

    # Check if translation was successful (not None)
    if translated_text is not None:
        return translated_text.lower()
    else:
        return text
```

```python
# Creating a function to just convert the text to lowercase
def lowercase(text):
  return text.lower()
```

```python
# Apply lowercase function to convert the data to lowercase

order_line['dim_category_translated'] = order_line['dim_category_translated'].apply(lowercase)
```

```python
order_line['dim_category_translated'].nunique()
```

```python
order_line['dim_category_translated'].unique()
```

Since some of the values need corrections. Make a list of the same and replace with the correct values.

```python
replacements = {
    "breizh&#39;n&#39;roll": "breizh n roll",
    "cooking": "cookings"
}

# Apply the replacements to the 'dim_category_translated' column

order_line['dim_category_translated']= order_line['dim_category_translated'].replace(replacements)
```

```python
order_line['dim_category_translated'].nunique()
```

```python
order_line['dim_category_translated'].value_counts()
```

```python
# Get unique values from the 'dim_category_translated' column
unique_categories = order_line['dim_category_translated'].unique()

# Sort the unique values alphabetically
sorted_categories = sorted(unique_categories)

# Print the sorted list of unique categories
for category in sorted_categories:
    print(category)
```

```python
# Defining categories from the data in column dim_category_translated

def categories(dim_category_translated):
    if dim_category_translated ==  'brunch' :
        return 'Main Course'
    if dim_category_translated ==   'cookings' :
        return 'Main Course'
    if dim_category_translated ==   'menu 25,00':
        return 'Main Course'
    elif dim_category_translated == 'breizh n roll' :
        return 'Desserts'
    elif dim_category_translated == 'buckwheat wrap' :
        return 'Appetizers'
    elif dim_category_translated == 'cold drinks' :
        return 'Soft Drinks'
    elif dim_category_translated == 'drinks' :
        return 'Wines'
    elif dim_category_translated == 'focaccia to take away' :
        return 'Starters'
    elif dim_category_translated == 'focaccia sur place' :
        return 'Cheeses'
    elif dim_category_translated == 'hot drinks' :
        return 'Hot Beverages'
    elif dim_category_translated == 'cafe / the' :
        return 'Beers'
    elif dim_category_translated == 'ice taste' :
        return 'Beers'
    elif dim_category_translated == 'martini 15 cl' :
        return 'Other Alcholic Drinks'
    elif dim_category_translated == 'champagne':
        return 'Other Alcholic Drinks'
    else:
        return 'others'
```

```python
order_line['categories'] = order_line['dim_category_translated'].apply(categories)
```
